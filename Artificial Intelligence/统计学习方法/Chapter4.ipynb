# 朴素贝叶斯法

朴素贝叶斯法是基于 **贝叶斯定理** 与 特征条件**独立假设** 的分类方法。



对于给定的数据集，首先基于特征条件独立假设学习输入/输出的联合概率分布；然后基于此模型，对给定的输入 x,利用贝叶斯定理求出后验概率最大的输入 y。

## 基本方法



朴素贝叶斯法通过学习先验概率分布

$$P(Y=c_k), k=1,2,..,K$$

及条件概率分布

$$P(X=x|Y=c_k)=P(X^{(1)}=x^{(1)},...,X^{(n)}=x^{(n)}|Y=c_k),k=1,2,..,K$$

根据联合概率分布密度函数：

$$P(X=x, Y=y) = P(Y=y|X=x)P(X=x) = P(X=x|Y=y)P(Y=y)$$

学习到联合概率分布 P(X,Y)。

条件概率分布有指数级数量的参数，其估计实际是不可行的。假设 X 可取值有 Sj 个，Y 可取值有 K 个，那么参数个数为 

$$K\prod_{j=1}^n S_j$$

取xj = S,参数个数为 $KS^n$ 当维数 n 很大时，就会发生维数灾难。

因此条件概率分布 $P(X=x | Y = c_k)$ 无法计算。

为了计算它，朴素贝叶斯法对它做了条件独立性假设


$$P(X=x|Y=c_k)=P(X_1=x_1,...,X_n=x_n|Y=c_k)=\prod_{j=1}^n P(X_j=x_j|Y=c_k)$$

后验概率为

$$P(Y=c_k|X=x)=\frac{P(X=x|Y=c_k)P(Y=c_k)}{\sum_k P(X=x|Y=c_k)P(Y=c_k)}$$

朴素贝叶斯分类器可以表示为：

$$y=f(x)=argmax \frac{\sum_j P(X_j=x_j|Y=c_k)P(Y=c_k)}{\sum_k P(Y=c_k)\prod_j P(X_j=x_j|Y=c_k)}$$

给定参数，找出概率最大的 c_k 出来，分母与c_k 无关去掉：

$$y=f(x)=argmax P(Y=c_k)\sum_j P(X_j=x_j|Y=c_k)$$


## 参数估计

朴素贝叶斯要学习的就是先验概率 $P(Y=c_k)$ 和条件概率 $P(X=x|Y=c_k)$ , 这两个估计用极大似然估计法进行：

先验概率：

$$P(Y=c_k)=\frac{\sum_{i=1}^N I(y_i = c_k)}{N}$$
